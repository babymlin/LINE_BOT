```python
import pandas as pd
import numpy as np
raw_df = pd.read_csv("log.csv", encoding="utf-8")
raw_df
```

```python
#取jsonPayload.message
import json
event_df = pd.DataFrame(columns=["userid", "event_type", "message_type", "message_text", "message_id", "timestamp"])
for row in raw_df["jsonPayload.message"].apply(json.loads):
    try:
        event = row.get("events")[0] #去event外面的list[]，且會有空值，設try/except
    except:
        pass
    if event.get("type") == "message":
        event_df = event_df.append({
            "userid" : event.get("source").get("userId"),
            "event_type" : event.get("type"),
            "message_type" : event.get("message").get("type"),
            "message_text" :event.get("message").get("text"),
            "message_id" : event.get("message").get("id"),
            "timestamp" : event.get("timestamp")},
            ignore_index = True)
    elif event.get("type") == "postback":
        event_df = event_df.append({
            "userid" : event.get("source").get("userId"),
            "event_type" : event.get("type"),
            "message_text" :event.get("postback").get("data"),
            "timestamp" : event.get("timestamp")},
            ignore_index = True)
    elif event.get("type") == "follow" or event.get("type") == "unfollow":
        event_df = event_df.append({
            "userid" : event.get("source").get("userId"),
            "event_type" : event.get("type"),
            "timestamp" : event.get("timestamp")},
            ignore_index = True)        
    else:
        event_df = event_df.append({
            "userid" : event.get("source").get("userId"),
            "event_type" : event.get("type"),
            "timestamp" : event.get("timestamp")},
            ignore_index = True)
event_df
```

```python
#設定LineBotApi準備向LINE取userId個人profile
from linebot import LineBotApi
with open("pass.txt", "r") as fp:
    pass_ = fp.read()
line_bot_api = LineBotApi(pass_.split(",")[0])

#依據jsonPayload中的userId向LINE取該userId的profile
user_info_df = pd.DataFrame(columns=["userid", "display_name", "picture_url", "status_message"])
for single in event_df["userid"].unique():
    user_info = line_bot_api.get_profile(single)
    try:
        user_info_df = user_info_df.append({
            "userid" : user_info.user_id,
            "display_name" : user_info.display_name,
            "picture_url" : user_info.picture_url,
            "status_message" : user_info.status_message}, #status_message會有空值，所以設定try/except
            ignore_index=True
        )
    except:
        pass
user_info_df    
```

```python
merge_df = pd.merge(event_df, user_info_df)
merge_df
```

```python
#timestamp轉換為現實時間
from datetime import datetime
merge_df["time"] = [datetime.fromtimestamp(val/1000).strftime('%Y-%m-%d %H:%M:%S') for val in merge_df["timestamp"]]
merge_df
```

```python
#將整理好的資料輸出成csv檔案
merge_df.to_csv(datetime.now().strftime("%Y%m%d") + ".csv", index=False)
```
